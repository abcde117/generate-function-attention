```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python


df = pd.read_csv("experiment_logs_om_ko.csv")

models = ["vit_po", "vit_pf", "vit_pok", "vit_pfk", "vit_base"]

plt.figure()

for m in models:
    sub = df[df["model"] == m]
    mean_df = sub.groupby("epoch")[["val_loss"]].mean().reset_index()
    plt.plot(mean_df["epoch"], mean_df["val_loss"], label=m)

plt.xlabel("Epoch")
plt.ylabel("Validation Loss")
plt.title("Validation Loss Comparison (mean over folds)")
plt.legend()
plt.show()
```
